```python
import json
import csv
import yaml
import pprint
```

```python
## load all of the predicted ORES topics
## this dataset has topic predictions for every article in Wikipedia

FILEPATH = '/Users/klogg/research_data/wmf_knowledge_graph/topics/topicsForAllWikipediaPages.csv'
ROWS= 2

with open(FILEPATH, newline='') as csvfile:
    CSVReader = csv.reader(csvfile, delimiter=',',escapechar='\\')
    for i,row in enumerate(CSVReader):
        print(', '.join(row))
        if i == ROWS:
            break
```

```python
## load all of the ground truth ORES topics
## this dataset only has topics for articles that were "hand labeled" by wikipedia editors

FILEPATH = '/Users/klogg/research_data/wmf_knowledge_graph/topics/labeled_enwiki_with_topics_metadata.json'

with open(FILEPATH) as json_file:
    for i,line in enumerate(json_file):
        pprint.pprint(json.loads(line))
        if i == ROWS:
            break
```

```python
FILEPATH = '/Users/klogg/research_data/wmf_knowledge_graph/topics/taxonomy.yaml'

with open(FILEPATH, 'r') as stream:
    try:
        pprint.pprint(yaml.safe_load(stream))
    except yaml.YAMLError as exc:
        print(exc)
```
