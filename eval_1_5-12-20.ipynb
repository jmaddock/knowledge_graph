```python
edge_df = pd.DataFrame()
for e in graph['tree_data']['edges']:
    try:
        df_dict = {
            'n1':e[0],
            'n2':e[1],
            'type':e[2]['type']
        }
        if 'depth' in e[2]:
            df_dict['depth'] = e[2]['depth']
        else:
            df_dict['depth'] = None
        if 'w' in e[2]:
            df_dict['w'] = e[2]['w']
        else:
            df_dict['w'] = None
    except:
        print(e)
        raise
    edge_df = edge_df.append(pd.DataFrame([df_dict]))
```
